```python
#정상 빈도수 x (성공)
import os
import pandas as pd

txt = []
bei = []

path = "C:\KISA-CISC2017-Malware-2nd[14]//" + '0//'
files = os.listdir(path)
type(files[0])
    
for file in files[:2]: 
    f = open(path + file, 'r')
   
    
    while True:
        line = f.readline().strip()
        if not line:
            break
        txt.append(line)    

    bei.append(txt)
    txt = []
f.close()


# print(bei)

#list를 DataFrame으로 (해결하기)
df = pd.DataFrame(bei)
df
```
